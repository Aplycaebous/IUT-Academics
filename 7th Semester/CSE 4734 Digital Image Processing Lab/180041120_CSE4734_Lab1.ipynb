#Task-1

**Explanation:** We define a function that takes the image path, and a threshold value. The function loops over the pixel intensity values and compares it with the threshold value. If the value is higher the output binary image's pixel value will be 1, otherwise 0.

```python
#Function which takes input path and threshold as input and returns a binary image
def to_binary(image_path, threshold):
  image=cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
  x, y = image.shape
  binary_image = np.zeros((x,y))
  for i in range(x):
    for j in range(y):
      if image[i][j]>threshold:
        binary_image[i][j] = 1
      else:
        binary_image[i][j] = 0
  return binary_image
```

```python
#Loading the images as grayscale images
bright = cv2.imread('/content/bright.jpg', cv2.IMREAD_GRAYSCALE)
gray = cv2.imread('/content/gray.jpg', cv2.IMREAD_GRAYSCALE)
washed_out = cv2.imread('/content/washed_out.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
#Defining 3 threshold values
THRESHOLD_1 = 256/4 #First Quarter
THRESHOLD_2 = 256/2 #Middle Value
THRESHOLD_3 = (256/4)*3 #Third Quarter
```

Now, we call the function on the loaded images using three threshold values

```python
binary_bright1 = to_binary("/content/bright.jpg", THRESHOLD_1)
binary_gray1 = to_binary("/content/gray.jpg", THRESHOLD_1)
binary_washed1 = to_binary("/content/washed_out.jpg", THRESHOLD_1)
```

```python
binary_bright2 = to_binary("/content/bright.jpg", THRESHOLD_2)
binary_gray2 = to_binary("/content/gray.jpg", THRESHOLD_2)
binary_washed2 = to_binary("/content/washed_out.jpg", THRESHOLD_2)
```

```python
binary_bright3 = to_binary("/content/bright.jpg", THRESHOLD_3)
binary_gray3 = to_binary("/content/gray.jpg", THRESHOLD_3)
binary_washed3 = to_binary("/content/washed_out.jpg", THRESHOLD_3)
```

```python
image=cv2.imread("/content/bright.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap = "gray")
plt.title("Grayscale Version of Bright Image")
plt.show()
```

```python
# calculate frequency of pixels in range 0-255
hist_bright = cv2.calcHist([bright],[0],None,[256],[0,256]) 
hist_gray = cv2.calcHist([gray],[0],None,[256],[0,256]) 
hist_washed = cv2.calcHist([washed_out],[0],None,[256],[0,256]) 
```

```python
plt.plot(hist_bright)
plt.title('Histogram of Bright Image')
plt.show()
```

**Explanation:** From the histogram, we can see a higher distribution of brighter values and can conclude that the image is a relatively birghter image.

```python
plt.plot(hist_gray)
plt.title('Histogram of Gray Image')
plt.show()
```

**Explanation:** From the histogram, we can tell the pixel values are almost evenly distributed. The image captured details in the whole spectrum.

```python
plt.plot(hist_washed)
plt.title('Histogram of Washed Out Image')
plt.show()
```

**Explanation:** From the histogram, we can see the distribution is rather smoothened which means it is a washed out image. The distribution is more dense towards higher values implying that it a relatively bright image.
